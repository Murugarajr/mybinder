# Vaex sandbox

Vaex is a high performance Python library for lazy Out-of-Core DataFrames (similar to Pandas), to visualize and explore big tabular datasets. It calculates statistics such as mean, sum, count, standard deviation etc, on an N-dimensional grid for more than a billion (10^9) samples/rows per second

```python
import vaex
```

```python
%%time
df = vaex.example()
df
```

```python
df.head(10)
```

```python
%%time
df.tail(10)
```

```python
df.x
```

```python
%%time
import numpy as np
np.sqrt(df.x**2 + df.y**2 + df.z**2)
```

```python
df['r'] = np.sqrt(df.x**2 + df.y**2 + df.z**2)
```

```python
%%time
df[['x', 'y', 'z', 'r']]
```

```python
df.select(df.x < 0)
df.evaluate(df.x, selection=True)
```

```python
df_negative = df[df.x < 0]
df_negative[['x', 'y', 'z', 'r']]
```

```python
df.count(), df.mean(df.x), df.mean(df.x, selection=True)
```

```python
df.column_count()
```

```python
df.info
```

```python
df.column_names
```

```python
df.description
```

```python
counts_x = df.count(binby=df.x, limits=[-10, 10], shape=64)
counts_x
```

```python
import matplotlib.pylab as plt
plt.plot(np.linspace(-10, 10, 64), counts_x)
plt.show()
```

```python
xycounts = df.count(binby=[df.x, df.y], limits=[[-10, 10], [-10, 20]], shape=(64, 128))
xycounts
```

```python
plt.imshow(xycounts.T, origin='lower', extent=[-10, 10, -10, 20])
plt.show()
```
